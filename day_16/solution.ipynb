```python
# Get adjecent positions
def get_adjacent(matrix, i, j):
    """Return a list of adjacent positions of (i, j) in a matrix"""
    L1, L2 = len(matrix), len(matrix[0])
    adjacent = []
    if i > 0:
        adjacent.append((i-1, j))
    if i < L1 - 1:
        adjacent.append((i+1, j))
    if j > 0:
        adjacent.append((i, j-1))
    if j < L2 - 1:
        adjacent.append((i, j+1))
    return adjacent
```

```python
def is_dead_end(maze, i, j):
    """Check if the given coordinates are a dead end in the maze."""
    adjacent = [position for position in get_adjacent(maze, i, j) if maze[position[0]][position[1]] in '#X']
    if(len(adjacent) == 3):
        return True
    return False
```

```python
# Define the unvisited set: dictionary with positions as keys and values as the cost to reach that position.
unvisited = defaultdict(lambda: 1e9)
for i in range(len(maze)):
    for j in range(len(maze[0])):
        if maze[i][j] in '.E':
            unvisited[(i, j)] = 1e9
        elif maze[i][j] in 'S':
            print(f'Starting position: ({i}, {j})')
            unvisited[(i, j)] = 0
# Sort the unvisited set by cost
sorted_unvisited = dict(sorted(unvisited.items(), key=lambda item: item[1]))
# Re-assign the unvisited set
unvisited = sorted_unvisited
```

```python
# Dijkstra's algorithm
# direction = 'East' # Initial direction
while True:
    # Current position: (i, j) with minimum cost to reach it.
    i, j = list(unvisited.keys())[0]
    if maze[i][j] == 'E':
        break
    adjacent = [position for position in get_adjacent(maze, i, j) if (maze[position[0]][position[1]] in '.E' and position in unvisited)]
    # print(f'Current position: ({i}, {j}), Cost to reach: {unvisited[(i, j)]}\n')
    for position in adjacent:
        # Distance from adjacent position to current position.
        # A change in direction costs 1001, otherwise 1.
        dx, dy = position[0] - i, position[1] - j
        if dx == 0 and dy == 1:
            new_direction = 'East'
        elif dx == 0 and dy == -1:
            new_direction = 'West'
        elif dx == 1 and dy == 0:
            new_direction = 'South'
        elif dx == -1 and dy == 0:
            new_direction = 'North'
        if new_direction != direction_dict[(i, j)][0]:
            distance = 1001
        else:
            distance = 1
        # Update the cost to reach the adjacent position.
        cost = min(unvisited[position], unvisited[(i, j)] + distance)
        unvisited[position] = cost
        direction_dict[position] = [new_direction, cost]  if direction_dict[(i, j)] != None else direction_dict[position]

        cost_dict[position].append([new_direction, unvisited[(i, j)] + distance])

    # Add the current position to the visited set.
    visited[(i, j)] = unvisited[(i, j)]

    # Remove the current position from the unvisited set (defaultdict).
    del unvisited[(i, j)]


    # Re-sort the unvisited set by cost
    sorted_unvisited = dict(sorted(unvisited.items(), key=lambda item: item[1]))
    # Re-assign this to the unvisited set.
    unvisited = sorted_unvisited
    # print(f'Unvisited costs: {unvisited}')

# The cost to reach the exit.
unvisited[(i, j)]
```

```python
# Loop which starts in the end position and looks for adjacent positions within the visited set.
# The each iteration will mark subsequent positions with a number.
# Start position will be marked with 0.
# Positions adjacent to the start position will be marked with 1.
# Positions adjacent to the positions marked with 1 will be marked with 2.
# This will build a map of the path from the end to the start.

# Assuming the start and end positions are known
start = (139, 1)  # Example start position
end = (1, 139)    # Example end position

# Define the map of the path
path_map = defaultdict(lambda: 1e9)
path_map[end] = 0

# Define the current number
number = 0

# Define the current positions
current_positions = [end]

# Loop to build the path map
while True:
    # Break the loop if the start position is reached
    if start in current_positions:
        break
    # Increment the number
    number += 1
    # Define the new positions
    new_positions = []
    # Loop through the current positions
    for i, j in current_positions:
        # Get the adjacent positions
        adjacent = [position for position in get_adjacent(maze, i, j) if (maze[position[0]][position[1]] in '.S' and path_map[position] == 1e9 and position in visited)]
        # Loop through the adjacent positions
        for position in adjacent:
            # Update the path map
            path_map[position] = number
            # Add the position to the new positions
            new_positions.append(position)
    # Update the current positions
    current_positions = new_positions
```

```python
# Now we start from the start position and look for adjacent positions with decreasing numbers.
# If at any point, there's two or more adjacent positions with the same number, we consider both of them.
# I.e., we consider the path to split into two, and both paths are valid.
# This should create ANOTHER path map, but this time from the start to the end.
# We continue this until the end position is reached.
# This will give us all possible paths from the start to the end.

# Loop which starts in the end position and looks for adjacent positions within the visited set.
# The each iteration will mark subsequent positions with a number.
# Start position will be marked with 0.
# Positions adjacent to the start position will be marked with 1.
# Positions adjacent to the positions marked with 1 will be marked with 2.
# This will build a map of the path from the end to the start.

# Define the current number
number = path_map[start]

# Define the current positions
current_positions = [start]

# Auxiliary map, where the paths will be marked
maze_path = maze.copy()

part2 = 1

vistited_set = set()

# Loop to build the path map
while True:
    # Decrement the number
    number -= 1
    # Define the new positions
    new_positions = []
    # Loop through the current positions
    for i, j in current_positions:
        # Get the adjacent positions
        adjacent = [position for position in get_adjacent(maze, i, j) if (maze[position[0]][position[1]] in '.E' and path_map[position] == number)]
        print(f'Current position: ({i}, {j}), Adjacent: {adjacent}')
        t_adj = [position for position in get_adjacent(maze, i, j) if (maze[position[0]][position[1]] in '.E' and position in visited)]
        print(f'Adjacent number(s): {[path_map[position] for position in t_adj]}')
        # Loop through the adjacent positions
        for position in adjacent:
            # Update the maze path map
            maze_path[position[0]] = maze_path[position[0]][:position[1]] + 'P' + maze_path[position[0]][position[1]+1:]

            # Add to the visited set
            len_t1 = len(vistited_set)
            vistited_set.add(position)
            len_t2 = len(vistited_set)

            # Increment the part 2 counter if the position WAS ADDED to the visited set
            if len_t1 != len_t2:
                part2 += 1

            # Add the position to the new positions
            new_positions.append(position)
    # Update the current positions
    current_positions = new_positions
    # Break the loop if the start position is reached
    if end in current_positions:
        break
print(f'Part 2: {part2}')
```

```python
from collections import defaultdict
from heapq import heappop, heappush

grid = {i+j*1j: c for i,r in enumerate(open('input.txt'))
                  for j,c in enumerate(r) if c != '#'}

start, = (p for p in grid if grid[p] == 'S')

seen = []
best = 1e9
dist = defaultdict(lambda: 1e9)
todo = [(0, t:=0, start, 1j, [start])]

while todo:
    val, _, pos, dir, path = heappop(todo)

    if val > dist[pos, dir]: continue
    else: dist[pos, dir] = val

    if grid[pos] == 'E' and val <= best:
        seen += path
        best = val

    for r, v in (1, 1), (+1j, 1001), (-1j, 1001):
        v, t, p, d = val+v, t+1, pos + dir*r, dir*r
        if p not in grid: continue
        heappush(todo, (v, t, p, d, path + [p]))


print(best, len(set(seen)))
```

```python
# replace positions in 'seen' with 'P' in the maze
# Note that these positions are represented as complex numbers, where the real part is the row and the imaginary part is the column.
maze_seen = maze.copy()
for position in seen:
    i, j = int(position.real), int(position.imag)
    maze_seen[i] = maze_seen[i][:j] + 'P' + maze_seen[i][j+1:]
for line in maze_seen:
    print(line)
```
