Part 1

```python
with open('input.txt') as f:
    string = f.read().strip()

n_ws = sum(int(string[i]) for i in range(1, len(string), 2))
len_string = sum(int(string[i]) for i in range(len(string)))
storage = [[int(string[i]), i // 2] if i % 2 == 0 else [int(string[i]), -1] for i in range(len(string))]

print(f'storage: {storage}')
print(f'len_string: {len_string}, n_ws: {n_ws}')
print(f'len_string - n_ws: {len_string - n_ws}')

stop = len_string - n_ws
```

```python
p1, n_spots_to_fill, i = 0, 0, 0

def part1(p1, n_spots_to_fill, i, storage, stop):
    for st_index, (n_ocurrences, value) in enumerate(storage):
        if value != -1:
            for _ in range(n_ocurrences):
                p1 += i * value
                i += 1
                if i == stop:
                    return p1
        else:
            n_spots_to_fill += n_ocurrences

        while n_spots_to_fill != 0:
            if storage[-1][0] != 0 and storage[-1][1] != -1:
                removable_spots = min(storage[-1][0], n_spots_to_fill)
                n_spots_to_fill -= removable_spots
                storage[st_index][0] -= removable_spots
                storage[-1][0] -= removable_spots

                for _ in range(removable_spots):
                    p1 += i * storage[-1][1]
                    i += 1
                    if i == stop:
                        return p1
            else:
                storage.pop(-1)
    return p1

p1 = part1(p1, n_spots_to_fill, i, storage, stop)
print(p1)
```

```python
def evolve(storage):
    # 'i_r' - index of the element to be moved ("reverse"); 'i_f' - index of the free space ("forward").
    L = len(storage)
    for i_r in range(L-1, 0, -1): # "0" 'cause we don't need to move the first element.
        if(storage[i_r][2] == 'M' or storage[i_r][1] == -1): continue # Testing 'i_r' - element already moved, or empty space.
        else:
            for i_f in range(1, L): # "1" 'cause we don't need to move the first element.
                if(i_r > i_f and storage[i_f][1] == -1 and storage[i_f][0] >= storage[i_r][0]): # If there's enough space.
                    storage[i_f][0] -= storage[i_r][0]
                    T = storage[i_r][1] # Save the value to be moved.
                    storage[i_r][1] = -1; storage[i_r][2] = '-' # Clear the space.
                    storage.insert(i_f, [storage[i_r][0], T, 'M'])
                    return evolve(storage)
            storage[i_r][2] = 'M' # If there's no space, mark the element as moved (anyway).
    return storage
```

```python
storage_vec = evolve([[int(string[i]), i // 2, 'N'] if i % 2 == 0 else [int(string[i]), -1, '-'] for i in range(len(string))])

i, p2 = 0, 0
for pos in storage_vec:
    if pos[1] != -1:
        p2 += sum((i + k) * pos[1] for k in range(pos[0]))
    i += pos[0]
print(p2)
# If I have time, I should look to see how other people did this, as I believe my solution is quite cumbersome.
```
